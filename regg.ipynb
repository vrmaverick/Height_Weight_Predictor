```python
def loss_func(m,b,data): #definig loss function
  total_error = 0
  for i in range(len(data)):
    x = data.iloc[i,0]
    y= data.iloc[i,1]
    total_error += (y-(m*x+b))**2
  return total_error / float(len(data))
```

```python
def gradient_descent(m,b,data,learning_rate):
  gradient_m = 0
  gradient_b = 0
  n= len(data)
  for i in range(n):
    x= data.iloc[i].Height
    y= data.iloc[i].Weight
    gradient_m += (-2/n)*x*(y-(m*x + b))
    gradient_m += (-2/n)*(y-(m*x + b))
  m -=(gradient_m *learning_rate)
  b -=(gradient_b * learning_rate)
  return m,b
```

```python
def linear_reggrssion(epochs ,data ,m =0,b=0,learning_rate=0.0001): #set default parameters

  print(f"Starting Linear Reggression with rate learning_rate {learning_rate} and epoches {epochs} .")

  for i in range (epochs):

    if i % 50 == 0:
      print(f"epoch : {i} ")#everypoint pe print karega

    m,b = gradient_descent(m,b,data,learning_rate) #calling functions
  return m,b
```

```python
epochs = 500
m,b =linear_reggrssion(epochs , df)
print(f' m = {m},b = {b}')

```
